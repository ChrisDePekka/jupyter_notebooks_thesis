```python
a = main_df[['duration2', "Deelnemer_ID", 'condition', 'ID_row', 'Ik heb de taak snel uitgevoerd']]
```

```python
print(a)
```

```python
a['Ik heb de taak snel uitgevoerd'] = a['Ik heb de taak snel uitgevoerd'].astype(int)
a['Deelnemer_ID'] = a['Ik heb de taak snel uitgevoerd'].astype(int)
a['condition'] = a['Ik heb de taak snel uitgevoerd'].astype(int)
```

```python
replacerid = {'NASGRM': 1, 'Only NA': 2,
            'Only SGRM': 3}
replacerdeel = {'Deelnemer_A': 1, 'Deelnemer_B': 2,
            'Deelnemer_C': 3}

# Select String Columns
cols = a.columns[a.dtypes == 'object']
# Replace Values in those Columns
a[cols] = a[cols].replace(replacerid)
a[cols] = a[cols].replace(replacerdeel)
```

```python
print(a.dtypes)
```

```python
pg.mixed_anova(dv='Ik heb de taak snel uitgevoerd', between='Deelnemer_ID', within='condition', subject='ID_row', data= a)
```

```python
aov = pg.mixed_anova(dv='Ik heb de taak snel uitgevoerd', between='Deelnemer_ID', within='condition', subject='ID_row', data= a)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import mixedlm

model = mixedlm("duration2 ~ condition", data=a, groups="Deelnemer_ID", re_formula="1")
result = model.fit()
print(result.summary())

# Explanation: https://vsni.co.uk/blogs/anova-or-LMM 
```
